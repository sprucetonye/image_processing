```python
colors = [
    [
        [0, 0, 255], #blue
        [0, 255, 0], #green 
    ], 
    [
        [255, 0, 0], #red
        [255, 255, 0] #yellow
    ]
]

print(np.array(colors).shape)
plt.imshow(colors)
start_row = 0
for row in colors:
    start_col = 0.25
    for color in row:
        plt.text(start_col, start_row, str(color))
        start_col += 1
    start_row += 1
```
